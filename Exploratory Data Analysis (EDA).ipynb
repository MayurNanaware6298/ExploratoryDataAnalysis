# EDA Project :

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
# Read CSV file
df = pd.read_csv('World_Army_Dataset.csv')
```

```python
df.head()
```

```python
# Above Population is in thousands so this function converts it single unit population
def population_correction(pop):
    return pop*1000
```

```python
df['Population'] = df['pop2022'].apply(population_correction)
```

```python
df['Population'] = df['Population'].astype('int')
```

```python
df.drop(['pop2022'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# This Function is to create percentage of perticular columns
def per_of_pop(mil,pop):
    return (mil/pop)*100
```

```python
df['total%'] = df.apply(lambda x:per_of_pop(x['total'],x['Population']),axis=1)
```

```python
df.head()
```

```python
df.info()
```

### Exploratory Data Analysis (EDA) :

```python
sns.set(rc={'figure.figsize':(20,10)})
a = sns.barplot(x='country',y='activeDuty',data=df.sort_values('activeDuty',ascending=False).head(10))

a.set_ylabel('Active Pesonels')
plt.show()
```

*Top 3 Largest active military countries are China, India and United States*

```python
df.columns
```

```python
b = sns.barplot(x='country',y='paramilitary',data=df.sort_values('paramilitary',ascending=False).head(10))

b.set_ylabel('Paramilitary Personels')
plt.show()
```

*From obove graph we can conclude North Korea, Vietnam and South Korea have highest number of paramilitary personels*

```python
df.columns
```

```python
c = sns.barplot(x='country', y='reserves',data=df.sort_values('reserves',ascending=False).head(10))

c.set_ylabel('Reserve Personels')
plt.show()
```

*From Above graph we can see Vietnam, South Korea and Russia have most number of reserve forces*

```python
df.columns
```

```python
d = sns.barplot(x='country',y='total%',data=df.sort_values('total%',ascending=False).head(10))

for bars in d.containers:
    d.bar_label(bars)
plt.show()
```

*From above graph we can conclude that in North Korea, South Korea and Cuba  have most percentage of population in miltary*

```python
sns.barplot(x='country',y='total', data=df.sort_values('total',ascending=False).head(10))
plt.show()
```

*From above graph we see Vietnam, North Korea and South Korea have highest number of military*
